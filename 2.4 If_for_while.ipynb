## 2.4 조건문과 반복문

```python
import numpy as np
```

### 2.4.1 IF문

```python
x = 10

if x == 1:
    print("x = 1")
else:
    print("x != 1")
```

```python
x = 10

if x == 1:
     print("x는 1 입니다.")
elif x == 2:
     print("x는 2 입니다.")
elif x == 3:
     print("x는 3 입니다.")
else:
     print("x는 4 이상 입니다.")
```

### 2.4.2 FOR/WHILE문

```python
for i in range(5): # range(5) = [0,1,2,3,4] list
     print(i)
```

```python
i = 0
while i < 5:
     print(i)
     i+=1 # i = i+1
```

### 2.4.3 break, continue

```python
for i in range(5):
     if i == 3:
          continue
     print(i)
```

```python
for i in range(5):
     if i == 3:
          break
     print(i)
```

### 2.4.4 try & except문

```python
x = [1,2,3,"s",4,5]

for i in x:
     try:
          y = i + 1
          print(y)
     except:
          pass
```

### 2.4.5 조건 반복문

```python
Z = np.random.randint(0,2, (10,3))

for z in Z:
     cond = sum(z)
     if cond == 3:
          print("1등 입니다.")
     elif cond == 0:
          print("2등 입니다.")
     elif (cond == 2) and (z[0] == 1):
          print("3등 입니다.")
     else:
          print("꽝 입니다.")
```
